```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import load_boston
from sklearn.model_selection import cross_val_score
from sklearn.impute import SimpleImputer
```

```python
boston = load_boston()
```

```python
boston.data.shape
```

```python
total_data_numbers = boston.data.size
total_data_samples = boston.data.shape[0]
total_data_features = boston.data.shape[1]
```

```python
rng = np.random.RandomState(0)
```

```python
X_missing = boston.data.copy()
```

```python
missing_rate=0.5
missing_data_numbers = int(total_data_numbers*missing_rate)
missing_samples_indexes=rng.randint(0,total_data_samples,missing_data_numbers)
missing_features_indexes=rng.randint(0,total_data_features,missing_data_numbers)
```

```python
X_missing[missing_samples_indexes,missing_features_indexes]=np.nan
X_missing = pd.DataFrame(X_missing)
#X_missing
```

```python
sort_null_indexes = np.argsort(X_missing.isnull().sum()).values
```

```python
isnull_column = X_missing.iloc[:,sort_null_indexes[0]].isnull()
```

```python
is_not_null_column = X_missing.iloc[:,sort_null_indexes[0]].notnull()
```

```python
X_missing = pd.DataFrame(X_missing)
```

```python
for i in np.argsort(X_missing.isnull().sum()).values:
    
    X_missing_i = X_missing.copy()
    X_missing_i = X_missing_i.iloc[0:,X_missing_i.columns!=i]
    X_new_missing = pd.concat([X_missing_i,pd.DataFrame(boston.target)],axis=1)
    
    is_null_rolls = X_missing.iloc[:,i].isnull()
    #Y_test = X_missing.iloc[:,i][is_null_rolls]

    
    is_not_null_rolls = X_missing.iloc[:,i].notnull()
    Ytrain = X_missing.iloc[:,i][is_not_null_rolls]
    
    X_new_missing_0 = SimpleImputer(missing_values=np.nan, 
                  strategy='constant',
                  fill_value=0
                 ).fit_transform(X_new_missing)
    X_new_missing_0 = pd.DataFrame(X_new_missing_0)
    
    X_new_missing_train = X_new_missing_0.loc[is_not_null_rolls,:]

    X_new_missing_test = X_new_missing_0.loc[is_null_rolls,:]

    Ypredict = RandomForestRegressor(n_estimators=100
                               ).fit(X_new_missing_train,Ytrain
                                    ).predict(X_new_missing_test)
    X_missing.loc[is_null_rolls,i] = Ypredict

  
```

```python
X_missing
```

```python
X = [boston.data,X_missing]
scores=[]
for x in X:
    rfc = RandomForestRegressor(random_state=0,n_estimators=100)
    score = cross_val_score(rfc,x,boston.target,scoring='neg_mean_squared_error',cv=5).mean()
    scores.append(score)
```
