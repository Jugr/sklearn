```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
```

```python
rng = np.random.RandomState(1)#随机数种子
```

```python
#rng.rand(80,1) #0-1间的x个随机数
#因为决策树不允许输入一维是数，要用reshape(-1,1)升维
```

```python
X = np.sort(5 * rng.rand(80,1), axis=0)#排序
# X
```

```python
y = np.sin(X).ravel()#降1维
#y.shape   (80,)
```

```python
y[::5] += 3 * (0.5 - rng.rand(16))
```

```python
plt.figure()
plt.scatter(X,y,s=20,edgecolors="black",c="darkorange")
```

```python
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(X, y)
regr_2.fit(X, y)
```

```python
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]#升1维
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)
```

```python
plt.figure()
plt.scatter(X, y, s=20, edgecolor="black",c="darkorange", label="data")
plt.plot(X_test, y_1, color="cornflowerblue",label="max_depth=2", linewidth=2)
plt.plot(X_test, y_2, color="yellowgreen", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
```
